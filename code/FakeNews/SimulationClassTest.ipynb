```python
from Simulation import Simulation
from News import News
import networkx as nx
%load_ext autoreload
%autoreload 2
```

```python
# News for simulation

sensation_news_1 = 0
sensation_news_2 = 0
decay_parameter_news_1 = 0
decay_parameter_news_2 = 0
n1 = News(1,sensation_news_1,decay_parameter_news_1)
n2 = News(2,sensation_news_2,decay_parameter_news_2)
news = {n1.name:n1, n2.name:n2}
```

```python
# Parameters of simulation

number_agents = 5
threshold = 0.2
independence = 0.1
initial_active_agents = {}
simulation_time = 5
```

```python
# Creates instance of simulation class

simulation = Simulation(number_agents, threshold, independence, news, simulation_time, initial_active_agents)
```

```python
# Draws the graph

options = { 'node_color': 'green',   'node_size': 1000,    'width': 1, 'font_size' : 20}  
nx.draw(simulation.world.graph, pos=nx.circular_layout(simulation.world.graph), with_labels = True, **options)  #
```

```python
# Runs the simulation

simulation.run_simulation()
```
